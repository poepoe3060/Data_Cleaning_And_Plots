###Global API

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize=(12, 6))

plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1) # row, column, pannel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['x^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])


plt.xlabel('X')
plt.ylabel('X Squared')
```

### Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N)) ** 2
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

###Histograms

```python
values = np.random.randn(1000)
```

```python
plt.figure(figsize=(12, 6))

plt.hist(values , bins=100, alpha=0.8, histtype='bar', color='orange', edgecolor='green')

plt.xlim(xmin=-5, xmax=5)

plt.show()

fig.savefig('hist.png')
```

###Kernel Density Estimation

```python
from scipy import stats
```

```python
density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)
plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

###Combine Plots

```python
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8, density=1, histtype='bar', color='#FF7F00', edgecolor='green')

plt.plot(values2, density(values2), color='red', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)
```
